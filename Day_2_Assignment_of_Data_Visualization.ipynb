```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris["species"].value_counts()
```

```python
iris.head()
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
sns.FacetGrid(iris, hue="species",palette='husl',height=5)\
    .map(sns.regplot,"sepal_width","sepal_length")\
    .add_legend()
```

```python
iris.plot(kind='scatter',x="sepal_width",y="sepal_length")
```

```python
sns.jointplot(x='sepal_width',y="sepal_length",data=iris, height=5)
```

```python
sns.pairplot(iris,hue='species', palette="Set2",diag_kind='kde',height=3)
```
